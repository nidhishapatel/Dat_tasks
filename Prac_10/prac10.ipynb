```python
# STEP 1: Import Required Libraries (Analytical Environment)
# -----------------------------------------

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder


# STEP 2: Load Dataset
# ------------------------------------------

df = pd.read_csv("StudentPerformanceFactors.csv")

print("Dataset Loaded Successfully")
print(df.head())
```

```python
# =====================================================
# Step 3: Data Visualization (Before Transformation)
# =====================================================

# 1. Distribution of Exam Score
plt.figure()
plt.hist(df['Exam_Score'])
plt.title("Exam Score Distribution (Before Scaling)")
plt.xlabel("Exam Score")
plt.ylabel("Frequency")
plt.show()

# 2. Hours Studied vs Exam Score
plt.figure()
plt.scatter(df['Hours_Studied'], df['Exam_Score'])
plt.title("Hours Studied vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.show()
```

```python
# =====================================================
# Step 4: Convert Categorical Data into Numeric
# =====================================================

encoder = LabelEncoder()

cat_cols = [
    'Gender', 'Parental_Involvement', 'Access_to_Resources',
    'Extracurricular_Activities', 'Motivation_Level',
    'Internet_Access', 'Family_Income', 'Teacher_Quality',
    'School_Type', 'Peer_Influence', 'Learning_Disabilities',
    'Parental_Education_Level'
]

for col in cat_cols:
    df[col] = encoder.fit_transform(df[col])
```

```python
# =====================================================
# Step 5: Standardize Numerical Data
# =====================================================

scaler = StandardScaler()

num_cols = [
    'Hours_Studied', 'Attendance', 'Sleep_Hours',
    'Previous_Scores', 'Tutoring_Sessions',
    'Physical_Activity', 'Exam_Score'
]

df[num_cols] = scaler.fit_transform(df[num_cols])


# =====================================================
# Step 6: Visualization After Standardization
# =====================================================

# 1. Distribution after scaling
plt.figure()
plt.hist(df['Exam_Score'])
plt.title("Exam Score Distribution (After Scaling)")
plt.xlabel("Scaled Exam Score")
plt.ylabel("Frequency")
plt.show()

# 2. Scatter after scaling
plt.figure()
plt.scatter(df['Hours_Studied'], df['Exam_Score'])
plt.title("Scaled Hours Studied vs Exam Score")
plt.xlabel("Scaled Hours Studied")
plt.ylabel("Scaled Exam Score")
plt.show()
```

```python
# =====================================================
# Step 7: Save Processed Dataset
# =====================================================

df.to_csv("standardized_student_data.csv", index=False)

print("\nStandardized dataset saved successfully!")
```
