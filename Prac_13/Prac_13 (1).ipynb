```python
# load the dataset

import pandas as pd
df = pd.read_csv("StudentPerformanceFactors.csv")
print(df.head())
```

```python
# Encode Categorical Variables
df = pd.get_dummies(df, columns=['Gender', 'School_Type'], drop_first=True)

print(df.head())
```

```python
# ==============================
# Label Encoding (Corrected)
# ==============================

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Clean column names (IMPORTANT)
df.columns = df.columns.str.strip()

# Check available columns
print("Available columns:\n", df.columns)

# ==============================
# Encode ONLY categorical columns
# ==============================

# Find categorical (object) columns
categorical_cols = df.select_dtypes(include=['object']).columns

print("\nCategorical columns to encode:", categorical_cols)

# Apply Label Encoding only on those columns
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

# ==============================
# Final Output Check
# ==============================
print("\nData after encoding:\n", df.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Select numerical columns
num_cols = [
    'Hours_Studied', 'Attendance', 'Sleep_Hours', 
    'Previous_Scores', 'Internet_Access',
    'Tutoring_Sessions', 'Physical_Activity', 'Exam_Score'
]

scaler = StandardScaler()

df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
from sklearn.preprocessing import MinMaxScaler

# Same numerical columns
num_cols = [
    'Hours_Studied', 'Attendance', 'Sleep_Hours', 
    'Previous_Scores', 'Internet_Access',
    'Tutoring_Sessions', 'Physical_Activity', 'Exam_Score'
]

scaler = MinMaxScaler()

df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df_subset = df[[
    'Hours_Studied',
    'Attendance',
    'Previous_Scores',
    'Sleep_Hours',
    'Tutoring_Sessions'
]]
print(df_subset.head())
```

```python
# GRoup By gender
gender_groups = df.groupby('Gender_Male')

for gender, group in gender_groups:
    print(f"\nGender: {gender}")
    print(group.head())

```

```python
#Group By school _types
school_groups = df.groupby('School_Type_Public')

for school, group in school_groups:
    print(f"\nSchool Type: {school}")
    print(group.head())
```
