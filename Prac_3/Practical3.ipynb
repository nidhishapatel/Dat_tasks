```python
'''
 Analytical Suitability Check of Dataset
 This program checks whether the dataset is suitable for
 data analysis by examining structure, data types,
 missing values, consistency, and variable distribution.'''
# -------------------------------------------------------------


import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv("StudentPerformanceFactors.csv")

# Display first 5 rows to preview the dataset
print("----- Dataset Preview -----")
print(df.head())

# Ensure columns have appropriate data formats
print("\n----- Data Types -----")
print(df.dtypes)
```

```python
# STEP 3: Check Data Structure
# Verify if the dataset is structured (rows and columns)
print("\n----- Data Structure Check -----")

if isinstance(df, pd.DataFrame):
    print("The dataset is STRUCTURED data (organized in rows and columns).")
else:
    print("The dataset is NOT structured.")


# STEP 4: Check Dataset Organization
# Display number of rows and columns
print("\n----- Dataset Organization -----")
print("Dataset Shape (Rows, Columns):", df.shape)

# Display column names
print("\nColumn Names:")
print(df.columns)
```

```python
print("\n----- Data Categorization -----")

# Numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64'])
print("\nNumerical Columns:")
print(numerical_columns.columns)

# Categorical columns
categorical_columns = df.select_dtypes(include=['object', 'string'])
print("\nCategorical Columns:")
print(categorical_columns.columns)
```

```python
# ============================================================
# STEP 7: Check Missing Values
# ============================================================

print("\n----- Missing Values Check (Before Cleaning) -----")

missing_values = df.isnull().sum()
print(missing_values)

# Check if dataset is clean
if missing_values.sum() == 0:
    print("No missing values found. Dataset is clean.")
else:
    print("Dataset contains missing values. Cleaning required.")


# ============================================================
# STEP 8: Fill Missing Values
# ============================================================

# Fill numeric columns with mean
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Fill categorical columns with mode
cat_cols = df.select_dtypes(include=['object', 'string']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])


# ============================================================
# STEP 9: Verify Cleaning
# ============================================================

print("\n----- Missing Values After Cleaning -----")
print(df.isnull().sum())

# Final check
if df.isnull().sum().sum() == 0:
    print("All missing values successfully handled. Dataset is now clean.")
else:
    print("Some missing values still remain.")
```

```python
# STEP 10: Check Data Consistency using Statistical Summary
# Helps detect incorrect values or outliers
print("\n----- Statistical Summary -----")
print(df.describe())


# STEP 11: Check Category Distribution
# Useful for evaluating balance in categorical variables
print("\n----- Category Distribution Example (Gender) -----")

if 'Gender' in df.columns:
    print(df['Gender'].value_counts())
```

```python
# STEP 12: Create Data Categorization Summary Table
print("\n----- Data Categorization Summary -----")

summary = pd.DataFrame({
    "Column Name": df.columns,
    "Data Type": df.dtypes,
    "Category": ["Numerical" if str(dtype) in ['int64','float64'] else "Categorical"
                 for dtype in df.dtypes]
})

print(summary)


# STEP 11: Final Analytical Suitability Conclusion
print("\n----- Analytical Suitability Conclusion -----")

print("1. Dataset is structured in tabular format.")
print("2. Data contains both numerical and categorical variables.")
print("3. Missing values were checked.")
print("4. Statistical summary confirms value ranges and consistency.")
print("5. Dataset is suitable for data analysis, visualization, and machine learning.")
```
