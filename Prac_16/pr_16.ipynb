```python
# Step 1: Setup & Import
import pandas as pd

# Step 2: Load JSON File
# Load JSON file
df = pd.read_json("currencies.json")

print("--- Raw JSON Data ---")
print(df.head())
```

```python
# Step 3: Flatten Nested Structure

# Extract the currencies list and normalize it
# Flatten the nested "currencies" list
df_structured = pd.json_normalize(df['currencies'])

print("\n--- Step 3: Flattened Data ---")
print(df_structured.head())
```

```python
# Step 4: Handle Semi-Structured Lists (Explode)

# Convert countries list into rows
df_final = df_structured.explode('countries')

print("\n--- Step 4: Fully Structured Data ---")
print(df_final.head(10))
```

```python
# Step 5: Summarization & Analysis
# Basic Statistics
print("\n--- Statistics ---")
print(df_final['decimal_digits'].describe())

# Average Decimal Digits per Country
summary = df_final.groupby('countries')['decimal_digits'].mean()
print("\n--- Average Decimal Digits per Country ---")
print(summary)

# Count of Countries per Currency
country_count = df_structured[['code', 'countries']].copy()
country_count['Country_Count'] = country_count['countries'].apply(len)
print("\n--- Countries per Currency ---")
print(country_count[['code', 'Country_Count']])
```

```python
# Step 6: Export to CSV
df_final.to_csv("structured_currencies.csv", index=False)

print("\nFile 'structured_currencies.csv' is ready for Excel.")
```
