```python
# =====================================================
# DATA INTEGRATION
# =====================================================

import pandas as pd

# Load dataset
data = pd.read_csv("StudentPerformanceFactors.csv")

# Create two smaller datasets (simulate different sources)

# 📘 Academic-related data
academic_data = data[['Id', 'Hours_Studied', 'Attendance', 'Previous_Scores', 'Exam_Score']]

# 🎯 Personal & environmental data
activity_data = data[['Id', 'Extracurricular_Activities', 'Sleep_Hours', 'Motivation_Level', 
                      'Internet_Access', 'Tutoring_Sessions', 'Family_Income']]

# Display sample datasets
print("Academic Data Sample:")
print(academic_data.head())

print("\nActivity Data Sample:")
print(activity_data.head())

# 🔗 Integrate datasets using Id
integrated_data = pd.merge(
    academic_data,
    activity_data,
    on="Id"
)

print("\nIntegrated Dataset:")
print(integrated_data.head())
```

```python
# =====================================================
# PATTERN DISCOVERY
# =====================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv("StudentPerformanceFactors.csv")

# STEP 1: Statistical summary
print("Statistical Summary:")
print(data.describe())

# STEP 2: Exam Score distribution
plt.hist(data['Exam_Score'], bins=10)

plt.title("Exam Score Distribution")
plt.xlabel("Exam Score")
plt.ylabel("Number of Students")

plt.show()

# STEP 3: Hours Studied vs Exam Score
plt.scatter(data['Hours_Studied'], data['Exam_Score'])

plt.title("Hours Studied vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")

plt.show()

# STEP 4: Attendance vs Exam Score
plt.scatter(data['Attendance'], data['Exam_Score'])

plt.title("Attendance vs Exam Score")
plt.xlabel("Attendance")
plt.ylabel("Exam Score")

plt.show()

# STEP 5: Correlation analysis
correlation_matrix = data.corr(numeric_only=True)

print("\nCorrelation Matrix:")
print(correlation_matrix)

# STEP 6: Heatmap visualization
sns.heatmap(correlation_matrix, annot=True)

plt.title("Feature Correlation Heatmap")

plt.show()
```
