```python
# =====================================================
# PART 1: DATA ACCESS CONTROL
# =====================================================

import pandas as pd

# STEP 1: Load dataset
data = pd.read_csv("StudentPerformanceFactors.csv")

# STEP 2: Define allowed columns (restricted view)
allowed_columns = [
    'Id',
    'Gender',
    'Parental_Education_Level',
    'Family_Income',
    'School_Type'
]

# STEP 3: Create restricted dataset
restricted_data = data[allowed_columns]

# STEP 4: Display restricted data
print("Restricted Data (Limited Access):")
print(restricted_data.head())
```

```python
# =====================================================
# PART 2: DATA PRIVACY PROTECTION
# =====================================================

import pandas as pd

# Load dataset
data = pd.read_csv("StudentPerformanceFactors.csv")

# STEP 1: Create a copy of dataset
privacy_data = data.copy()

# STEP 2: Mask Id values (anonymization)
privacy_data['Id'] = privacy_data['Id'].apply(lambda x: "ID_#@*" + str(x)[-3:])

# STEP 3: (Optional) Mask other sensitive columns

# Mask Family Income
privacy_data['Family_Income'] = "Hidden"

# Mask Gender (optional)
privacy_data['Gender'] = "X"

# STEP 4: Display anonymized dataset
print("Anonymized Data:")
print(privacy_data.head())
```

```python
# =====================================================
# PART 3: DATA SECURITY
# =====================================================

import pandas as pd

# Load dataset
data = pd.read_csv("StudentPerformanceFactors.csv")

# STEP 1: Create secure backup
data.to_csv("secure_student_backup.csv", index=False)

# STEP 2: Confirmation message
print("Secure dataset backup created successfully")
```

```python
# =====================================================
# PART 4: DATA QUALITY STANDARDS
# =====================================================

import pandas as pd

# Load dataset
data = pd.read_csv("StudentPerformanceFactors.csv")

# STEP 1: Check missing values in each column
missing_values = data.isnull().sum()

print("Missing Values in Each Column:")
print(missing_values)

# STEP 2: Check duplicate records
duplicate_records = data.duplicated().sum()

print("\nNumber of Duplicate Records:", duplicate_records)

# STEP 3: Show columns with missing values only
print("\nColumns with Missing Values:")
print(missing_values[missing_values > 0])
```

```python
# =====================================================
# PART 5: DATA COMPLIANCE MONITORING
# =====================================================

import pandas as pd

# Load dataset
data = pd.read_csv("StudentPerformanceFactors.csv")

# STEP 1: Identify invalid attendance values
# Rule: Attendance must be between 0 and 100
invalid_attendance = data[
    (data['Attendance'] < 0) | (data['Attendance'] > 100)
]

# STEP 2: Display invalid records
print("Invalid Attendance Records:")
print(invalid_attendance)
```
