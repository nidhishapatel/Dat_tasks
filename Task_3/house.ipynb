```python
house_width = 11
right_width = 20
left_margin = 5

# ROOF
for i in range(1, house_width, 2):
    spaces = (house_width - i) // 2
    house = " " * spaces + "*" * i + " " * spaces
    print(" " * left_margin + house + "*" * right_width)

# TOP BORDER
print(" " * left_margin + "*" * house_width + "*" * right_width)

# WINDOW
print(" " * left_margin + "*   ***   *" + "*" * right_width)

# EMPTY WALL
print(" " * left_margin + "*         *" + "*" * right_width)

# DOOR (LEFT SIDE)
print(" " * left_margin + "*  ***    *" + "*" * right_width)
print(" " * left_margin + "*  * *    *" + "*" * right_width)
print(" " * left_margin + "*  ***    *" + "*" * right_width)

# BOTTOM BORDER
print(" " * left_margin + "*" * house_width + "*" * right_width)
```

```python
house_width = 11
right_width = 20

# ROOF (Fixed Slanted)
for i in range(1, house_width, 2):
    spaces = (house_width - i) // 2
    roof = " " * spaces + "*" * i
    print(roof + "*" * right_width)

# FULL TOP LINE
print("*" * house_width + "*" * right_width)

# BODY WITH WINDOW
print("*   ***   *" + "*" * right_width)

# EMPTY WALL
print("*         *" + "*" * right_width)

# DOOR (3 lines)
print("*    ***    *"[:house_width] + "*" * right_width)
print("*    * *    *"[:house_width] + "*" * right_width)
print("*    ***    *"[:house_width] + "*" * right_width)

# BOTTOM LINE
print("*" * house_width + "*" * right_width)
```
