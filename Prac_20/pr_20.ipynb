```python
# Step 1: Setup and Load Data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

# Load your dataset
df = pd.read_csv("StudentPerformanceFactors.csv")

# Preview
print("--- Dataset Preview ---")
print(df.head())
```

```python
# Step 2: Central Tendency (Mean, Median, Mode)
# Select numerical columns
num_cols = ['Hours_Studied', 'Attendance', 'Sleep_Hours',
            'Previous_Scores', 'Tutoring_Sessions',
            'Physical_Activity', 'Exam_Score']

print("\n--- Step 2: Central Tendency ---")

for col in num_cols:
    mean_val = df[col].mean()
    median_val = df[col].median()
    mode_val = df[col].mode()[0]

    print(f"\n{col}:")
    print(f"Mean: {round(mean_val,2)}")
    print(f"Median: {median_val}")
    print(f"Mode: {mode_val}")
```

```python
# Step 3: Dispersion (Range, Variance, Std Dev)

print("\n--- Step 3: Dispersion ---")

for col in num_cols:
    range_val = df[col].max() - df[col].min()
    variance_val = df[col].var()
    std_dev = df[col].std()

    print(f"\n{col}:")
    print(f"Range: {range_val}")
    print(f"Variance: {round(variance_val,2)}")
    print(f"Standard Deviation: {round(std_dev,2)}")
```

```python
# Step 4: Quartiles & Percentiles

print("\n--- Step 4: Quartiles ---")

for col in num_cols:
    quartiles = df[col].quantile([0.25, 0.5, 0.75])

    print(f"\n{col}:")
    print(quartiles)
```

```python
# Step 5: Box Plot (Outliers Detection)
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['Exam_Score'], color='lightgreen')

plt.title('Statistical Spread of Exam Scores')
plt.xlabel('Exam Score')
plt.show()
```
