```python
# Step 1: Setup & Load Dataset
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("StudentPerformanceFactors.csv")

print("--- Step 1: Basic Dataset ---")
print(df.head())
```

```python
# Step 2: Enhancement (Feature Engineering)

# Create meaningful new columns from your dataset
# Performance Index (combination of key factors)
df['Performance_Index'] = (
    df['Hours_Studied'] * 0.4 +
    df['Attendance'] * 0.3 +
    df['Previous_Scores'] * 0.3
)

# Study Efficiency (score per hour studied)
df['Study_Efficiency'] = df['Exam_Score'] / df['Hours_Studied']

# Lifestyle Score (sleep + physical activity)
df['Lifestyle_Score'] = df['Sleep_Hours'] + df['Physical_Activity']

print("\n--- Step 2: Enhanced Data ---")
print(df[['Hours_Studied', 'Attendance', 'Exam_Score',
          'Performance_Index', 'Study_Efficiency', 'Lifestyle_Score']].head())
```

```python
# Step 3: Quality Improvement (Scaling / Normalization)

# Normalize important numeric columns (0 to 1 scale)

cols_to_scale = ['Hours_Studied', 'Attendance', 'Previous_Scores', 'Exam_Score']

for col in cols_to_scale:
    df[col + '_Scaled'] = (
        (df[col] - df[col].min()) /
        (df[col].max() - df[col].min())
    )

print("\n--- Step 3: Scaled Data ---")
print(df[[col + '_Scaled' for col in cols_to_scale]].head())
```

```python
# Step 4: Restructuring (Pivot Table)

# Compare Exam Score across Gender & School Type

pivot_df = df.pivot_table(
    index='Gender',
    columns='School_Type',
    values='Exam_Score',
    aggfunc='mean'
)

print("\n--- Step 4: Pivot Table ---")
print(pivot_df)
```

```python
# Step 5: Aggregation (Summary Table)

# High-level performance summary
summary = df.groupby('Gender').agg({
    'Exam_Score': 'mean',
    'Hours_Studied': 'mean',
    'Attendance': 'mean',
    'Performance_Index': 'mean'
}).rename(columns={
    'Exam_Score': 'Avg_Score',
    'Hours_Studied': 'Avg_Study_Hours',
    'Attendance': 'Avg_Attendance'
})

print("\n--- Step 5: Final Aggregated Summary ---")
print(summary)
```
