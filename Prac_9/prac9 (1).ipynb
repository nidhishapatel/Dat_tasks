```python
# =====================================================
# PRACTICAL 9: DATA ACQUISITION & INTEGRATION
# =====================================================

# Step 1: Import Libraries
import pandas as pd

# =====================================================
# Step 2: Load Multiple Datasets
# =====================================================

# Dataset 1 (Academic Data)
academic = pd.read_csv("academic_data.csv")

# Dataset 2 (Student Personal Data)
personal = pd.read_csv("student_info.csv")

print("Academic Data:")
print(academic.head())

print("\nPersonal Data:")
print(personal.head())


# =====================================================
# Step 3: Check Dataset Structure
# =====================================================

print("\nAcademic Dataset Info:")
print(academic.info())

print("\nPersonal Dataset Info:")
print(personal.info())


# =====================================================
# Step 4: Integrate Datasets (Merge)
# =====================================================

# Merge using common column 'Id'
merged_data = pd.merge(academic, personal, on="Id")

print("\nMerged Dataset:")
print(merged_data.head())


# =====================================================
# Step 5: Save Integrated Dataset
# =====================================================

merged_data.to_csv("student_complete_data.csv", index=False)

print("\nMerged dataset saved successfully!")


# =====================================================
# Step 6: Perform Basic Analysis
# =====================================================

# Example: Total Academic Effort
merged_data["Total_Effort"] = (
    merged_data["Hours_Studied"] +
    merged_data["Tutoring_Sessions"] +
    merged_data["Physical_Activity"]
)

print("\nDataset with Total Effort:")
print(merged_data.head())
```
