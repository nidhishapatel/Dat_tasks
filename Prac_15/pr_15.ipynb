```python
# import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Load Tabular Data
df = pd.read_csv("StudentPerformanceFactors.csv")
```

```python
# 2. Structured Summarization
print("--- Dataset Snapshot ---")
print(df.head(10)) 
```

```python
print("\n--- Summary Statistics ---")
print(df.describe())
```

```python
# Step 5: Structured Summarization (Key Insights)

# Average Exam Score
print("\n---- Average Exam Score ----")
print(df['Exam_Score'].mean())

# Average Hours Studied
print("\n---- Average Hours Studied ----")
print(df['Hours_Studied'].mean())
```

```python
# Step 6: Categorical Interpretation

# Parental Involvement vs Exam Score
print("\n---- Parental Involvement vs Exam Score ----")
print(df.groupby('Parental_Involvement')['Exam_Score'].mean())

# Motivation Level vs Exam Score
print("\n---- Motivation Level vs Exam Score ----")
print(df.groupby('Motivation_Level')['Exam_Score'].mean())

# Gender vs Exam Score
print("\n---- Gender vs Exam Score ----")
print(df.groupby('Gender')['Exam_Score'].mean())
```

```python
# Step 7: Visual Interpretation (Box Plot)

# Exam Score Distribution
plt.figure(figsize=(6,4))
df.boxplot(column='Exam_Score')
plt.title("Distribution of Exam Scores")
plt.ylabel("Marks")
plt.show()

# Hours Studied Distribution
plt.figure(figsize=(6,4))
df.boxplot(column='Hours_Studied')
plt.title("Distribution of Hours Studied")
plt.ylabel("Hours")
plt.show()
```
